```python
import pandas as pd
import numpy as np
df = pd.read_csv("Cricketer_Stats.csv")
df
```

```python
df.duplicated()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
len(df)
```

```python
df.loc[df['Runs'] < 0, 'Runs'] = np.nan
df.loc[df['Matches_Played'] < 0, 'Matches_Played'] = np.nan
df.loc[df['Wickets'] < 0, 'Wickets'] = np.nan
df.loc[df['Batting_Avg'] < 0, 'Batting_Avg'] = np.nan
df.loc[df['Bowling_Avg'] < 0, 'Bowling_Avg'] = np.nan
df
```

```python
df['Country'].unique()
df['Role'].unique()
```

```python
df['Role'] = df['Role'].replace({
    'all-rounder': 'All-Rounder',  
    'batsman': 'Batsman',
    'bowler': 'Bowler'
})
df
```

```python
df['Runs'] = df['Runs'].fillna(round(df['Runs'].mean()))
df['Matches_Played'] = df['Matches_Played'].fillna(round(df['Matches_Played'].median()))
df['Wickets'] = df['Wickets'].fillna(round(df['Wickets'].median()))
df['Batting_Avg'] = df['Batting_Avg'].fillna(round(df['Batting_Avg'].mean()))
df['Bowling_Avg'] = df['Bowling_Avg'].fillna(round(df['Bowling_Avg'].mean()))
df
```
