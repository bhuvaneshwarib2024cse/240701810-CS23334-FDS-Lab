```python
# Step 1: Import libraries and load dataset
import numpy as np
import pandas as pd
df = pd.read_csv("Hotel_Dataset - Hotel_Dataset.csv")
df
```

```python
# Step 2: Check for duplicates
df.duplicated()
```

```python
# Step 3: Check dataset info
df.info()
```

```python
# Step 4: Drop duplicate rows
df.drop_duplicates(inplace=True)
df
```

```python
len(df)
```

```python
# Step 5: Drop unnecessary column 'Age_Group.1'
df.drop(['Age_Group.1'], axis=1, inplace=True)
df
```

```python
# Step 6: Correct negative or impossible values
df.loc[df['CustomerID'] < 0, 'CustomerID'] = np.nan
df.loc[df['Bill'] < 0, 'Bill'] = np.nan
df.loc[df['EstimatedSalary'] < 0, 'EstimatedSalary'] = np.nan
df.loc[(df['NoOfPax'] < 1) | (df['NoOfPax'] > 20), 'NoOfPax'] = np.nan
df.loc[(df['Rating(1-5)'] < 1) | (df['Rating(1-5)'] > 5), 'Rating(1-5)'] = np.nan
df
```

```python
# Step 7: Standardize categorical columns
df['Hotel'] = df['Hotel'].replace(['Ibys'], 'Ibis')
df['FoodPreference'] = df['FoodPreference'].replace(['veg', 'Vegetarian', 'Veg'], 'Veg')
df['FoodPreference'] = df['FoodPreference'].replace(['non-Veg', 'non-veg', 'Non-Veg'], 'Non-Veg')
df
```

```python
# Step 8: Fill missing or corrected values
df['EstimatedSalary'] = df['EstimatedSalary'].fillna(round(df['EstimatedSalary'].mean()))
df['NoOfPax'] = df['NoOfPax'].fillna(round(df['NoOfPax'].median()))
df['Rating(1-5)'] = df['Rating(1-5)'].fillna(round(df['Rating(1-5)'].median()))
df['Bill'] = df['Bill'].fillna(round(df['Bill'].mean()))
df
```

```python
# Step 9: Reset index
df.reset_index(drop=True, inplace=True)
df
```
